```python
tr['is_peak_hour'] = tr['Time_Orderd'].apply(is_peak_hour)
peak_orders = tr[tr['is_peak_hour']].groupby('Order_Date').size().reset_index(name='peak_order_volume')
non_peak_orders = tr[~tr['is_peak_hour']].groupby('Order_Date').size().reset_index(name='non_peak_order_volume')
tr
```

```python
tr['Order_Date'] = pd.to_datetime(tr['Order_Date'])
```

```python
peak_orders = tr[tr['is_peak_hour'] == True].groupby('Order_Date').size().reset_index(name='peak_order_volume')
non_peak_orders = tr[tr['is_peak_hour'] == False].groupby('Order_Date').size().reset_index(name='non_peak_order_volume')
```

```python
order_volumes = pd.merge(peak_orders, non_peak_orders, on='Order_Date', how='outer').fillna(0)
order_volumes = order_volumes.sort_values(by='Order_Date')
print(order_volumes)
```

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

model_peak = ARIMA(order_volumes['peak_order_volume'], order=(5, 1, 0))
model_peak_fit = model_peak.fit()
forecast_peak = model_peak_fit.forecast(steps=7)

model_non_peak = ARIMA(order_volumes['non_peak_order_volume'], order=(5, 1, 0))
model_non_peak_fit = model_non_peak.fit()
forecast_non_peak = model_non_peak_fit.forecast(steps=7)

plt.figure(figsize=(12, 6))
plt.plot(order_volumes['Order_Date'], order_volumes['peak_order_volume'], label='Peak Order Volume')
plt.plot(order_volumes['Order_Date'], order_volumes['non_peak_order_volume'], label='Non-Peak Order Volume')
plt.plot(pd.date_range(start=order_volumes['Order_Date'].iloc[-1], periods=8, freq='D')[1:], forecast_peak, label='Forecast Peak', linestyle='--')
plt.plot(pd.date_range(start=order_volumes['Order_Date'].iloc[-1], periods=8, freq='D')[1:], forecast_non_peak, label='Forecast Non-Peak', linestyle='--')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Order Volume')
plt.title('Order Volume Forecast for Peak and Non-Peak Hours')
plt.show()
```

```python
mean_peak = order_volumes['peak_order_volume'].mean()
std_peak = order_volumes['peak_order_volume'].std()

mean_non_peak = order_volumes['non_peak_order_volume'].mean()
std_non_peak = order_volumes['non_peak_order_volume'].std()

def calculate_buffer_percentage(mean, std, z_score=1.96):
    buffer = z_score * std
    buffer_percentage = (buffer / mean) * 100
    return buffer_percentage

buffer_percentage_peak = calculate_buffer_percentage(mean_peak, std_peak)
buffer_percentage_non_peak = calculate_buffer_percentage(mean_non_peak, std_non_peak)

print(f"Buffer percentage for peak hours: {buffer_percentage_peak:.2f}%")
print(f"Buffer percentage for non-peak hours: {buffer_percentage_non_peak:.2f}%")
```

```python
import math
orders_per_rider_per_hour = 10
hours_per_shift = 8

def estimate_riders(forecast_orders, orders_per_rider_per_hour, hours_per_shift, buffer_percentage):
    total_orders = forecast_orders.sum()
    riders_needed = total_orders / (orders_per_rider_per_hour * hours_per_shift)
    buffer_riders = riders_needed * (buffer_percentage / 100)
    return math.ceil(riders_needed), math.ceil(buffer_riders)

# Example
forecast_peak = order_volumes['peak_order_volume'].tail(7)   
forecast_non_peak = order_volumes['non_peak_order_volume'].tail(7)  


riders_peak, buffer_riders_peak = estimate_riders(forecast_peak, orders_per_rider_per_hour, hours_per_shift, buffer_percentage_peak)
riders_non_peak, buffer_riders_non_peak = estimate_riders(forecast_non_peak, orders_per_rider_per_hour, hours_per_shift, buffer_percentage_non_peak)

print(f"Estimated riders needed for peak hours: {riders_peak}")
print(f"Buffer riders needed for peak hours: {buffer_riders_peak}")
print(f"Estimated riders needed for non-peak hours: {riders_non_peak}")
print(f"Buffer riders needed for non-peak hours: {buffer_riders_non_peak}")
```
